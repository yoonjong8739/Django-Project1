```python
import pandas as pd
import numpy as np

import re
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
#from time import sleep
import time
```

```python
book_title_list = []
book_author_list = []
book_price_list = []
book_star_list = []
book_cate_list = []
book_isbn_list = []
book_tag_list = []
book_img_list = []
book_recommend_list = []
```

```python
for i in range(1, 51):
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get("http://www.yes24.com/24/Category/BestSeller?CategoryNumber=017&sumgb=06&AO=2&PageNumber={}".format(i))
    res = driver.page_source
    soup = BeautifulSoup(res, 'lxml')

    book_links = soup.select('td.image > div.goodsImgW > a')
    book_link_list = []
    driver.quit()
    
    # 미리보기 링크 제거
    for book_link in book_links:
        if "javascript:openEBookPreview" in book_link['href']:
            continue
        else:
            book_url = "http://www.yes24.com" + book_link['href']
            book_link_list.append(book_url)

    driver = webdriver.Chrome(ChromeDriverManager().install())
    for i in range(len(book_link_list)):
        url2 = book_link_list[i]
        driver.get(url2)
        res2 = driver.page_source
        soup2 = BeautifulSoup(res2, 'lxml')

        #book_title = soup2.find("h2", {"class":"gd_name"}).get_text()  # 책 제목
        #book_title_list.append(book_title)

        try:
            book_title = soup2.find("h2", {"class":"gd_name"}).get_text()  # 책 제목
        except AttributeError:
            book_title_list.append("NULL")
        else:
            book_title_list.append(book_title)

        try:
            book_author = soup2.find("span", {"class":"gd_auth"}).get_text().replace("\n","").replace("\r","").rstrip().lstrip()  # 저자
        except AttributeError:
            book_author_list.append("NULL")
        else:
            book_author_list.append(book_author)

        try:
            book_price = soup2.find("span", {"class":"nor_price"}).get_text()  # 가격
        except AttributeError:
            book_price_list.append("NULL")
        else:
            book_price_list.append(book_price)

        try:
            book_star = soup2.select_one('div.review_starWrap > div.gd_rating > .yes_b').get_text()  # 별점
        except AttributeError:
            book_star_list.append("NULL")
        else:
            book_star_list.append(book_star)

        try:
            book_cate = soup2.select_one("dd>.yesAlertLi>li").get_text().replace("\n","").replace(" ","").replace("\xa0","")  # 카테고리 분류
        except AttributeError:
            book_cate_list.append("NULL")
        else:
            book_cate_list.append(book_cate)

        try:
            book_isbn = soup2.find("div", {"class":"yesTb"}).get_text().replace("\n", "").replace("\r","")[-13:]  # ISBN13
        except AttributeError:
            book_isbn_list.append("NULL")
        else:
            if re.compile("[^0-9]").match(book_isbn):
                book_isbn = book_isbn.replace(book_isbn, "0"*13)
                book_isbn_list.append(book_isbn)
                #print(book_isbn)
            else:
                #print(book_isbn)
                book_isbn_list.append(book_isbn)

        try:
            book_tag = soup2.find("div", {"class":"gd_tagArea"}).get_text().replace("\n","")  # 태그
        except AttributeError:
            book_tag_list.append("NULL")
        else:
            book_tag_list.append(book_tag)

        try:
            book_img = soup2.find("img", {"class":"gImg"})['src']  # 책 이미지 url
        except AttributeError:
            book_img_list.append("NULL")
        else:
            book_img_list.append(book_img)

        recommends = soup2.select('#nomiBoxRoolGrp_buyNCateGoodsWrap > .clearfix > #recommend_goods_area > .goods_info')
        recommend_list = []
        for recommend in recommends:
            recom_books = recommend.find('p').get_text()
            recommend_list.append(recom_books)

        book_recommend_list.append(recommend_list)
    driver.quit()
```

```python
print(book_title_list)
print(book_author_list)
print(book_price_list)
print(book_star_list)
print(book_cate_list)
print(book_isbn_list)
print(book_tag_list)
print(book_img_list)
print(book_recommend_list)
```

```python
yes24 = pd.DataFrame({"ISBN" : book_isbn_list,
                 "book_name" : book_title_list,
                 "author" : book_author_list,
                 "price" : book_price_list,
                 "star" : book_star_list,
                 "category" : book_cate_list,
                 "tag" : book_tag_list,
                 "with_buy_book" : book_recommend_list,
                 "img" : book_img_list})
yes24
```
